```python
from pandas import *
from numpy import *
from sklearn import *
from sklearn.metrics import *
from matplotlib.pyplot import *
# Importing the dataset
dataset = read_csv('kd.csv')
dataset.isnull().any()
ls1=['age','blood_pressure','sg','albumin','su','bgr','bu','Serum_creatinine','sod','pot','Haemoglobin']
mean=dataset.mean(axis=0)
for i in ls1:
    dataset[i]=dataset[i].fillna(mean[i])
dataset.iloc[:,16]=dataset.iloc[:,16].fillna(method='pad')
dataset.iloc[:,17]=dataset.iloc[:,17].fillna(method='pad')
dataset.iloc[:,18]=dataset.iloc[:,18].fillna(method='pad')
dataset['classification']=dataset['classification'].replace('ckd\t','ckd')
dataset['classification'].unique()
dataset
```

```python
dataset=dataset.dropna(axis=0)
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
ls=[6,7,8,9,19,20,21,22,23,24,25]
for i in ls:   
    dataset.iloc[:,i]=enc.fit_transform(dataset.iloc[:,i])
cor=dataset.iloc[:,1:].corr()
from seaborn import *
cmap=diverging_palette(220,10,as_cmap=True)
heatmap(cor,cmap=cmap,annot=False)
dataset['classification'].unique()
dataset
```

```python
#feature selection
columns = np.full((cor.shape[0],), True, dtype=bool)
for i in range(cor.shape[0]):
    for j in range(i+1, cor.shape[0]):
        if cor.iloc[i,j] >= 0.5:
            if columns[j]:
                columns[j] = False
selected_columns = cor.columns[columns]
selected_columns =selected_columns.append(Index(['pus_cell']))
x=dataset[selected_columns]
y=dataset['classification']
area=[]
```

```python
selected_columns.shape
```

```python
def auc_scorer( pred, y1, model): # Helper function to plot the ROC curve
    if model=='RF':
        fpr, tpr, rocnauc= roc_curve(y1,pred)
        print('accuracy score:',accuracy_score(y1,pred))
    elif model=='svm(linear)':
        fpr, tpr, rocnauc = roc_curve(y1, pred)
        print('accuracy score:',accuracy_score(y1,pred))
    elif model=='svm(kernal)':
        fpr, tpr, rocnauc = roc_curve(y1, pred)
        print('accuracy score:',accuracy_score(y1,pred))
    elif model=='knn':
        fpr, tpr, rocnauc = roc_curve(y1, pred)
        print('accuracy score:',accuracy_score(y1,pred))
    elif model=='decision':
        fpr, tpr, rocnauc = roc_curve(y1, pred)
        print('accuracy score:',accuracy_score(y1,pred))
    elif model=='logistic':
        fpr, tpr, rocnauc = roc_curve(y1, pred)
        print('accuracy score:',accuracy_score(y1,pred))
    elif model=='NB':
        fpr, tpr, rocnauc = roc_curve(y1, pred)
        print('accuracy score:',accuracy_score(y1,pred))
    roc_auc = auc(fpr, tpr)
    area.append(roc_auc)

    figure()    
    plot(fpr, tpr, label='ROC curve from '+model+' model (area = %0.3f)' % roc_auc)
    plot([0, 1], [0, 1], 'k--')
    xlim([0.0, 1.0])
    ylim([0.0, 1.05])
    xlabel('False Positive Rate')
    ylabel('True Positive Rate')
    title('ROC Curve')
    legend(loc="lower right")
    show()

    return fpr,tpr,roc_auc
```

```python
 #data analysis using distribution plot
fig = figure(figsize = (20, 25))
j = 0
for i in x.columns:
    subplot(6, 4, j+1)
    j += 1
    distplot(x[i][dataset['classification']==0], color='g', label = 'chronic kidney disease')
    distplot(x[i][dataset['classification']==1], color='r', label = 'Not chronic kidney disease')
    legend(loc='best')
fig.suptitle('survival analysis(chronic kidney disease)')
fig.tight_layout()
fig.subplots_adjust(top=0.95)
show()
```

```python
#splitting 
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.375,random_state=0)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#x_train = sc.fit_transform(x_train)
#x_test = sc.transform(x_test)
```

```python
#naive Bayes
from sklearn.naive_bayes import GaussianNB 
clas1= GaussianNB()
clas1.fit(x_train,y_train)
y_pred1=clas1.predict(x_test)
from sklearn.metrics import confusion_matrix
cm1=confusion_matrix(y_test,y_pred1)
auc_scorer( y_pred1, y_test, 'NB')
a1=precision_recall_fscore_support(y_test,y_pred1)
print('confusion matrix:\n',cm1)
clas1.predict(array([53.000000,90.000000,1.020,2.0,0.0,0,1,0,114.000000,3.70000,0,0]).reshape(1,-1))
```

```python
#random forest 
from sklearn.ensemble import RandomForestClassifier 
clas2= RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
clas2.fit(x_train,y_train)
y_pred2=clas2.predict(x_test)
cm2=confusion_matrix(y_test,y_pred2)
auc_scorer(y_pred2, y_test, 'RF')
a2=precision_recall_fscore_support(y_test,y_pred2)
print('confusion matrix:\n',cm2)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
clas3= DecisionTreeClassifier(criterion='entropy',random_state=10)
clas3.fit(x_train,y_train)
y_pred3=clas3.predict(x_test)
cm3=confusion_matrix(y_test,y_pred3)
auc_scorer(y_pred3, y_test, 'decision')
a3=precision_recall_fscore_support(y_test,y_pred3)
print('confusion matrix:\n',cm3)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
clas4 = LogisticRegression(random_state = 110)
clas4.fit(x_train, y_train)
y_pred4 = clas4.predict(x_test)
cm4=confusion_matrix(y_test,y_pred4)
auc_scorer(y_pred4, y_test, 'logistic')
a4=precision_recall_fscore_support(y_test,y_pred4)
print('confusion matrix:\n',cm4)
#clas4.predict(array([48,80,1.025,1]).reshape(1,-1))
```

```python
#svm linear
from sklearn.svm import SVC
clas5 = SVC(kernel = 'linear', random_state = 10)
clas5.fit(x_train, y_train)
y_pred5 = clas5.predict(x_test)
cm5=confusion_matrix(y_test,y_pred5)
auc_scorer(y_pred5, y_test, 'svm(linear)')
a5=precision_recall_fscore_support(y_test,y_pred5)
print('confusion matrix:\n',cm5)
#clas6.predict(array([48,80,1.025,0,1,0,1]).reshape(1,-1))
```

```python
#svm kernal
from sklearn.svm import SVC
clas6 = SVC(kernel = 'rbf', random_state = 100)
clas6.fit(x_train, y_train)
y_pred6 = clas6.predict(x_test)
cm6=confusion_matrix(y_test,y_pred6)
auc_scorer(y_pred6, y_test, 'svm(kernal)')
a6=precision_recall_fscore_support(y_test,y_pred6)
print('confusion matrix:\n',cm6)
#clas6.predict(array([57,1.025,1,1,0,0,0]).reshape(1,-1))
```

```python
#knn classifier
from sklearn.neighbors import KNeighborsClassifier
clas7 = KNeighborsClassifier(n_neighbors=10,metric='minkowski',p=2)
clas7.fit(x_train, y_train)
y_pred7 = clas7.predict(x_test)
cm7=confusion_matrix(y_test,y_pred7)
auc_scorer(y_pred7, y_test, 'knn')
a7=precision_recall_fscore_support(y_test,y_pred7)
print('confusion matrix:\n',cm7)
#clas7.predict(array([57,80,1.025,0,0,1,0,0,135,4.7,0,1]).reshape(1,-1))
```
